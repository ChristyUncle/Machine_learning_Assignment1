Before split the dataset, need to re-balance both loan status class 0 and 1, I used the resample method, to add more data in minority class (loan status = 1)

````python
#Imbalanced Data, therefore needs to balance both classes
from sklearn.utils import resample

#create two different dataframe of majority and minority class
data_majority = data[(data["Loan Status"]==0)]
data_minority = data[(data["Loan Status"]==1)]

# upsample minority class
data_minority_upsampled = resample(data_minority,
                                 replace=True,    # sample with replacement
                                 n_samples= 9058, # to match majority class
                                 random_state=0)

# Combine majority class with upsampled minority class
data = pd.concat([data_minority_upsampled, data_majority])
````

````python
#check is there any missing values
missing_values = data.isnull().sum()

print("Missing Values in Each Column:")
print(missing_values)
````

#Reasoning and interpretation for ElasticNet
L1 ratio indicates the optimal mix between L1(LASSO)and L2(Ridge) regularization in the Elastic Net, thereby establishing a compromise between including all features and sparsity. Our optimal L1 ratio is 0.5, which indicates an equal mix of L1 and L2 regularization.

Alpha is the penalty term that indicates the amount of shrinkage (or constraint) that will be implemented in the equation. If the alpha is set to 0 in either the LASSO or Ridge model, the estimation results will be equivalent to the linear regression model. A larger alpha value penalizes the optimization function. A lower alpha allows for more complex models with larger coefficients. The best alpha and lambda values are those values that minimize the cross-validation error.

Our optimal alpha is 0.0004, which means that the model is less penalized for having large coefficients. That is, a low alpha indicates weaker regulization, making the model less constrained. #Pitfall
L1 Regularization
L1 Regularization tends to shrink coefficients of less important features to  precisely zero and leading to sparse model. Also L1 regulation may also leads to instability by select one variable over another, especially when predictor variables are highly correlated may cause L1 regularization encounter the problem with multicollinearity.

L2 Regularization
Drawbacks of L2 Regularization are that Ridge regession retains all features in the model, estimating non-zero coefficients to all of them. That is, L2 does not automatically perfrom feature selection. Moreover, Ridge performs poorly when it comes to sparse data, as many feature values may be zero. This can lead to suboptimal results in case that only a subset of features has explanatory power.

Furthermore, a lower alpha value allows for a more complex model that fits the training data more narrowly. However, an alpha value that is too low can lead to overfitting.

Therefore, one needs to be careful in the selection and tuning of regularization parameters to catch the right balance between bias and variance to optimalize the models performance.
